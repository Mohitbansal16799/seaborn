```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")
data = sns.load_dataset("tips")
```

```python
data
```

```python
data.head()
```

# plot a scatter plot between total bill and tip

```python

sns.relplot(x = "total_bill",y = "tip",data = data)
```

# the Hue parameter 

```python
sns.relplot(x="total_bill",y = "tip",hue = "smoker",data = data)
```

# the style parameter

```python
sns.relplot(x = "total_bill",y = "tip",hue = "smoker",style = "sex",data = data)
```

# the size parameter

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",style="sex",size="size",data=data)
```

#    # scatter plot function

```python
sns.scatterplot(x = "total_bill",y = "tip",data = data,hue = "smoker")
```

# Categorical scatter plot

```python
#x axis = categorical data
#y axis = numerical data
#there is a two type of categorical plot 
#1 strip plot
#2 swarm plot
```

```python
#strip plot
sns.catplot(x = "day",y = "tip",data = data,kind = "strip")
```

```python
#swarm plot
sns.catplot(x = "day",y = "tip",kind = "swarm",data = data,hue = "sex")
```

# Boxplot 

```python
#univariate analysis
```

```python
sns.boxplot(data["tip"])
```

```python
sns.boxplot(data["total_bill"])
```

```python
sns.boxplot(x = data["day"],y = data["tip"])
```

```python
sns.catplot(x = "day",y ="tip",hue = "sex",data = data,kind = "box")
```

#  violinplot

```python
sns.violinplot(x = data["day"],y = data["tip"])
```

# Barplot

```python
sns.catplot(x="smoker",y="total_bill",kind ="bar",data=data)
```

```python
sns.catplot(x="smoker",y="total_bill",kind ="bar",data=data,hue = "sex")
```

# countplot

```python
sns.catplot(x = "sex",kind = "count",hue = "smoker",data = data)
```

# Heatmap

```python
flight = sns.load_dataset("flights")
```

```python
flight.head()
```

```python
x = flight.pivot_table(index = "year",columns = "month",values = "passengers",aggfunc = "sum")
```

```python
sns.heatmap(x)
```

# clustermap

```python
sns.clustermap(x)
```
